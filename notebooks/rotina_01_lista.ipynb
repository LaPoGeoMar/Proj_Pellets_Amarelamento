## 01. Abrindo a imagem de pellet

```python
#--- Passo 1: Instalação do pacote Scikit-Image ---#
"""conda install skimage"""

#--- Passo 2: Importando o pacote Scikit-Image ---#

import skimage
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
#--- Passo 3: Colocando as fotos que serão utilizadas em uma lista ---#

from skimage import data
from skimage import io

lista_fotos = os.listdir('fotos_pellets_cp')
#fotos_pellets_cp = pasta de arquivo para testes

#quantidade de fotos na lista 
quantidade_fotos = len(lista_fotos)
print(quantidade_fotos)
print(lista_fotos)
```

```python
#--- Passo 4: Abrindo as imagens ---#

#Abrindo uma foto apenas

from skimage import io
foto = io.imread('fotos_pellets_cp/cl1_p1_psul_deixa1_g0005.jpg')

print(plt.imshow(foto))
```
