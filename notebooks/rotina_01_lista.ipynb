## 01. Abrindo a imagem de pellet

```python
# --- Passo 1: Instalação do pacote Scikit-Image ---#
"""conda install skimage"""

# --- Passo 2: Importando o pacote Scikit-Image ---#

import os

import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
# --- Passo 4: Abrindo as imagens ---#

# Abrindo uma foto apenas

from skimage import io

foto = io.imread("fotos_pellets_cp/cl1_p1_psul_deixa1_g0005.jpg")

print(plt.imshow(foto))
```
